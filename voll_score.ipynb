### Źródło danych

Dane o wynikach zawodników pochodzą z https://en.volleyballworld.com/en/vnl/2019/men/statistics a pobrane w zostały z https://www.kaggle.com/kerneler/starter-fibv-2020-statistics-eb7dfa73-a/data . Są to dane z wstępnej fazy ligi narodów w 2019 roku.

Dane są podzielona na 6 kategorii staystyk odnoszących się do:
1. Ataku
2. Bloku
3. Obrony
4. Przyjęcia
5. Zagrywki
6. Rozegrania

W kazdej z tych kategorii przypiszemy zawodnikowi pewną ilość punktów, a nastepnie określimy jego finalny wynik sumując rezultat poszczególnych kategorii.

Metryki odpowiednich kategorii będą się nazywać:  
1. Attack Score  
2. Block Score  
3. Dig Score  
4. Receive Score  
5. Serve Score  
6. Set Score  

Parametry do obliczenia każdej z tych metryk są uznaniowe i bazują na doświadczeniu badacza które nie jest duże.

Końcową mętrykę będącą sumą powyższych nazywamy **Total Score**

### Uwagi

Metryka Total Score sprawdza się tylko do porównania zawodników którzy rozegrali porównywalną ilość setów/akcji ponieważ nie uwzglednia ich ilości.

```python
import pandas as pd
```

```python
men_best_attackers = pd.read_csv("./FIBV_2020_Statistics_PreliminaryRound/Men_Statistics/men_best_attackers.csv")
men_best_blockers  = pd.read_csv("./FIBV_2020_Statistics_PreliminaryRound/Men_Statistics/men_best_blockers.csv")
men_best_diggers   = pd.read_csv("./FIBV_2020_Statistics_PreliminaryRound/Men_Statistics/men_best_diggers.csv")
men_best_recievers = pd.read_csv("./FIBV_2020_Statistics_PreliminaryRound/Men_Statistics/men_best_recievers.csv")
men_best_servers   = pd.read_csv("./FIBV_2020_Statistics_PreliminaryRound/Men_Statistics/men_best_servers.csv")
men_best_setters   = pd.read_csv("./FIBV_2020_Statistics_PreliminaryRound/Men_Statistics/men_best_setters.csv")
```

### Attack Score

Do policzenia wyniku za ataki posłużymy się wzorem:

Attack Score = 0.8 * (ataki pomyślne) - (ataki błędne)

```python
def attack_score(row):
    return 0.8 * row['Spikes'] - row['Faults']

men_best_attackers['Attack Score'] = men_best_attackers.apply(attack_score, axis=1)

men_best_attackers.sort_values(by=['Attack Score'], ascending=False)
```

### Block Score

Do policzenia wyniku za bloki posłużymy się wzorem:

Block Score = 1.5 * (bloki pomyślne) - 0.25 * (bloki błędne) + 0.5 * (odbicia)

```python
def block_score(row):
    return 1.5 * row['Stuff Blocks'] - 0.3 * row['Faults'] + 0.5 * row['Rebounds']

men_best_blockers['Block Score'] = men_best_blockers.apply(block_score, axis=1)

men_best_blockers.sort_values(by=['Block Score'], ascending=False)
```

### Dig Score

Do policzenia wyniku za obrony posłużymy się wzorem:
    
Dig Score =  (obrony z przyjęciem) + 0.3 * obrony - 0.3 * (obrony błędne)

```python
def dig_score(row):
    return 1.0 * row['Receptions']  + 0.3 * row['Digs']  - 0.3 * row["Faults"]

men_best_diggers['Dig Score'] = men_best_diggers.apply(dig_score, axis=1)

men_best_diggers.sort_values(by=['Dig Score'], ascending=False)
```

### Recieve Score

Do policzenia wyniku za przyjęcia posłużymy się wzorem:

Recieve Score = 0.5 * (przyjęcia wyśmienite) + 0.2 * (przyjęcia) - 0.2 * (przyjęcia niepomyślne)

```python
def recieve_score(row):
    return 0.5 * row['Excellents'] + 0.2 * row['Serve receptions'] - 0.2 * row["Faults"]

men_best_recievers['Recieve Score'] = men_best_recievers.apply(recieve_score , axis=1)

men_best_recievers.sort_values(by=['Recieve Score'], ascending=False)
```

### Serve Score

Do policzenia wyniku za zagrywki posłużymy się wzorem:

Serve Score = (asy) + 0.2 * (zagrywki w pole) - (faule)

```python
def serve_score(row):
    return row['Aces'] + 0.2 * row['Hits'] - row["Faults"]

men_best_servers['Serve Score'] = men_best_servers.apply(serve_score, axis=1)

men_best_servers.sort_values(by=['Serve Score'], ascending=False)
```

### Set Score

Do policzenia wyniku za rozegrania posłużymy się wzorem:

Set Score = 0.12 * (zagrywki na maksymlanie 1 blokującego) + 0.6 * (zagrywki na 2 lub więcej blokujących) - faule

```python
def set_score(row):
    return 0.12 * row['Running sets']  + 0.06 * row['Still sets'] - row["Faults"]

men_best_setters['Set Score'] = men_best_setters.apply(set_score , axis=1)

men_best_setters.sort_values(by=['Set Score'], ascending=False)
```

### Total Score

Końcowy wynik zawodnika jest sumą jego wyniki z wszystkich sześciu kategorii.

```python
scores = men_best_attackers[['Name', 'ShirtNumber', 'Team']].copy()
scores = pd.merge(scores, men_best_attackers[['Team', 'ShirtNumber', 'Attack Score']], on=['Team', 'ShirtNumber'])
scores = pd.merge(scores, men_best_blockers[['Team', 'ShirtNumber', 'Block Score']], on=['Team', 'ShirtNumber'])
scores = pd.merge(scores, men_best_diggers[['Team', 'ShirtNumber', 'Dig Score']], on=['Team', 'ShirtNumber'])
scores = pd.merge(scores, men_best_recievers[['Team', 'ShirtNumber', 'Recieve Score']], on=['Team', 'ShirtNumber'])
scores = pd.merge(scores, men_best_servers[['Team', 'ShirtNumber', 'Serve Score']], on=['Team', 'ShirtNumber'])
scores = pd.merge(scores, men_best_setters[['Team', 'ShirtNumber', 'Set Score']], on=['Team', 'ShirtNumber'])

def total_score(row):
    return row['Attack Score'] + row['Block Score'] + row['Dig Score'] + row['Recieve Score'] + row['Serve Score'] + row['Set Score']

scores['Total Score'] = scores.apply(total_score, axis=1)
```

### Wyniki dla reprezentatów Polski

```python
scores[scores['Team'] == 'POL'].sort_values(by=['Total Score'], ascending=False)
```
